```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgBikang = cv2.imread('bikang3.png')
imgDadar = cv2.imread('dadar_gulung3.png')
imgKlepon = cv2.imread('klepon3.png')
imgMolen = cv2.imread('molen_pisang3.png')

img_rgbBikang = cv2.cvtColor(imgBikang, cv2.COLOR_BGR2RGB)
img_rgbDadar = cv2.cvtColor(imgDadar, cv2.COLOR_BGR2RGB)
img_rgbKlepon = cv2.cvtColor(imgKlepon, cv2.COLOR_BGR2RGB)
img_rgbMolen = cv2.cvtColor(imgMolen, cv2.COLOR_BGR2RGB)

grayBikang = cv2.cvtColor(img_rgbBikang, cv2.COLOR_RGB2GRAY)
grayDadar = cv2.cvtColor(img_rgbDadar, cv2.COLOR_RGB2GRAY)
grayKlepon = cv2.cvtColor(img_rgbKlepon, cv2.COLOR_RGB2GRAY)
grayMolen = cv2.cvtColor(img_rgbMolen, cv2.COLOR_RGB2GRAY)


plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(img_rgbBikang)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(grayBikang, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(img_rgbDadar)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(grayDadar, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(img_rgbKlepon)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(grayKlepon, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(img_rgbMolen)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(grayMolen, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

```

```python
gaussBikang = cv2.GaussianBlur(grayBikang, (5,5), 0)
medianBikang = cv2.medianBlur(grayBikang, 5)

gaussDadar = cv2.GaussianBlur(grayDadar, (5,5), 0)
medianDadar = cv2.medianBlur(grayDadar, 5)

gaussKlepon = cv2.GaussianBlur(grayKlepon, (5,5), 0)
medianKlepon = cv2.medianBlur(grayKlepon, 5)

gaussMolen = cv2.GaussianBlur(grayMolen, (5,5), 0)
medianMolen = cv2.medianBlur(grayMolen, 5)

#test
blurBikang = cv2.blur(grayBikang, (5,5))

plt.figure(figsize=(12,4))
plt.subplot(1,3,1);
plt.imshow(medianBikang, cmap='gray');
plt.title('Original Blur Gray Bikang');
plt.axis('off')

plt.subplot(1,3,2);
plt.imshow(gaussBikang, cmap='gray');
plt.title('Gaussian Blur Bikang');
plt.axis('off')

plt.subplot(1,3,3);
plt.imshow(medianBikang, cmap='gray');
plt.title('Median Blur Bikang');
plt.axis('off')

plt.figure(figsize=(12,4))

plt.subplot(1,3,1);
plt.imshow(grayDadar, cmap='gray');
plt.title('Original Gray Dadar');
plt.axis('off')

plt.subplot(1,3,2);
plt.imshow(gaussDadar, cmap='gray');
plt.title('Gaussian Blur Dadar');
plt.axis('off')

plt.subplot(1,3,3);
plt.imshow(medianDadar, cmap='gray');
plt.title('Median Blur Dadar');
plt.axis('off')

plt.figure(figsize=(12,4))

plt.subplot(1,3,1);
plt.imshow(grayKlepon, cmap='gray');
plt.title('Original Gray Klepon');
plt.axis('off')

plt.subplot(1,3,2);
plt.imshow(gaussKlepon, cmap='gray');
plt.title('Gaussian Blur Klepon');
plt.axis('off')

plt.subplot(1,3,3);
plt.imshow(medianKlepon, cmap='gray');
plt.title('Median Blur Klepon');
plt.axis('off')

plt.figure(figsize=(12,4))

plt.subplot(1,3,1);
plt.imshow(grayMolen, cmap='gray');
plt.title('Original Gray Molen');
plt.axis('off')

plt.subplot(1,3,2);
plt.imshow(gaussMolen, cmap='gray');
plt.title('Gaussian Blur Molen');
plt.axis('off')

plt.subplot(1,3,3);
plt.imshow(medianMolen, cmap='gray');
plt.title('Median Blur Molen');
plt.axis('off')
plt.show()
```

```python
kernel_sharp_basic = np.array([[0,-1,0],
                              [-1,5,-1],
                              [0,-1,0]])

kernel_sharp_mild = np.array([[0, -0.5, 0],
                              [-0.5, 3, -0.5],
                              [0, -0.5, 0]])

kernel_sharp_strong = np.array([[-1, -1, -1],
                                [-1, 9, -1],
                                [-1, -1, -1]])

kernel_sharp_custom = np.array([[1, -2, 1],
                                [-2, 5, -2],
                                [1, -2, 1]])

sharp_gaussBikang = cv2.filter2D(gaussBikang, -1, kernel_sharp_strong)
sharp_gaussBikang = cv2.filter2D(gaussBikang, -1, kernel_sharp_strong)

sharp_medianBikang = cv2.filter2D(medianBikang, -1, kernel_sharp_strong)
sharp_gaussianBikang = cv2.filter2D(gaussBikang, -1, kernel_sharp_strong)

sharp_medianDadar = cv2.filter2D(medianDadar, -1, kernel_sharp_strong)
sharp_gaussianDadar = cv2.filter2D(gaussDadar, -1, kernel_sharp_strong)

sharp_medianKlepon = cv2.filter2D(medianKlepon, -1, kernel_sharp_strong)
sharp_gaussianKlepon = cv2.filter2D(gaussKlepon, -1, kernel_sharp_strong)

sharp_medianMolen = cv2.filter2D(medianMolen, -1, kernel_sharp_strong)
sharp_gaussianMolen = cv2.filter2D(gaussMolen, -1, kernel_sharp_strong)


plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(sharp_medianBikang, cmap='gray')
plt.title("Sharpened median Bikang Image")
plt.axis('off')


plt.subplot(1,2,2)
plt.imshow(sharp_gaussianBikang, cmap='gray')
plt.title("Sharpened gaussian Bikang Image")
plt.axis('off')


plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(sharp_medianDadar, cmap='gray')
plt.title("Sharpened Median Dadar Image")
plt.axis('off')


plt.subplot(1,2,2)
plt.imshow(sharp_gaussianDadar, cmap='gray')
plt.title("Sharpened gaussian Dadar Bikang Image")
plt.axis('off')


plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(sharp_medianKlepon, cmap='gray')
plt.title("Sharpened Median Klepon Image")
plt.axis('off')


plt.subplot(1,2,2)
plt.imshow(sharp_gaussianKlepon, cmap='gray')
plt.title("Sharpened gaussian Klepon Image")
plt.axis('off')


plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(sharp_medianMolen, cmap='gray')
plt.title("Sharpened Median Molen Image")
plt.axis('off')


plt.subplot(1,2,2)
plt.imshow(sharp_gaussianMolen, cmap='gray')
plt.title("Sharpened gaussian Molen Image")
plt.axis('off')
plt.show()
```

```python
_, th0 = cv2.threshold(sharp_gaussianBikang, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
_, th1 = cv2.threshold(sharp_gaussianDadar, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
_, th2 = cv2.threshold(sharp_gaussianKlepon, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
_, th3 = cv2.threshold(sharp_gaussianMolen, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

_, th0_0 = cv2.threshold(sharp_medianBikang, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
_, th1_1 = cv2.threshold(sharp_medianDadar, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
_, th2_2 = cv2.threshold(sharp_medianKlepon, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
_, th3_3 = cv2.threshold(sharp_medianMolen, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)


th_inv0 = cv2.bitwise_not(th0)
th_inv1 = cv2.bitwise_not(th1)
th_inv2 = cv2.bitwise_not(th2)
th_inv3 = cv2.bitwise_not(th3)

th_inv0_0 = cv2.bitwise_not(th0_0)
th_inv1_1 = cv2.bitwise_not(th1_1)
th_inv2_2 = cv2.bitwise_not(th2_2)
th_inv3_3 = cv2.bitwise_not(th3_3)

plt.figure(figsize=(5,5))
plt.imshow(th_inv0, cmap='gray')
plt.title("Segmented Cells gaussian (Binary Mask with OTSU)")
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(th_inv1, cmap='gray')
plt.title("Segmented Cells gaussian (Binary Mask OTSU)")
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(th_inv2, cmap='gray')
plt.title("Segmented Cells gaussian (Binary Mask OTSU)")
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(th_inv3, cmap='gray')
plt.title("Segmented Cells gaussian (Binary Mask OTSU)")
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(th_inv0_0, cmap='gray')
plt.title("Segmented Cells median (Binary Mask OTSU)")
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(th_inv1_1, cmap='gray')
plt.title("Segmented Cells median (Binary Mask OTSU)")
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(th_inv2_2, cmap='gray')
plt.title("Segmented Cells median (Binary Mask OTSU)")
plt.axis('off')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(th_inv3_3, cmap='gray')
plt.title("Segmented Cells median (Binary Mask OTSU)")
plt.axis('off')
plt.show()
```
